```python
# from IPython.display import clear_output


# To print out Board

def display_board(board):
    clear_output()
    print('   '+'  '+'|'+'   '+'|'+' ')      
    print('   '+board[7]+' |'+' '+board[8]+' '+'|'+' '+board[9])
    print('   '+'  '+'|'+'   '+'|'+' ')   
    print('---------------')
    print('   '+'  '+'|'+'   '+'|'+' ')     
    print('   '+board[4]+' |'+' '+board[5]+' '+'|'+' '+board[6])
    print('   '+'  '+'|'+'   '+'|'+' ')   
    print('---------------')
    print('   '+'  '+'|'+'   '+'|'+' ')   
    print('   '+board[1]+' |'+' '+board[2]+' '+'|'+' '+board[3])
    print('   '+'  '+'|'+'   '+'|'+' ') 
   




  
        
        
    
            
   
```

```python
# Assign marker to the players!


def player_input():
    user_input='K'
    while user_input not in ['X','O']:
        user_input=input('Player1,choose your mark (X or O):')
        if user_input not in ['X','O']:
            print('I cant understand! Please enter correctly')
        else:
            player1=user_input
            
    
    if player1=='X':
        player2='O'
    else:
        player2='X'
    return (player1,player2)
            



    
    
        
    
```

```python
# Place marker on the Board

def place_marker(board, marker, position):
    board[position]=marker
    
    
```

```python
# check if someone won!


def win_check(board, mark):
    return(board[7]==board[8]==board[9]==mark or
           board[4]==board[5]==board[6]==mark or 
           board[1]==board[2]==board[3]==mark or 
           board[1]==board[4]==board[7]==mark or 
           board[2]==board[5]==board[8]==mark or 
           board[3]==board[6]==board[9]==mark or
           board[1]==board[5]==board[9]==mark or
           board[3]==board[5]==board[7]==mark)
        
```

```python
# Randomly pick which player will go first!

import random
def choose_first():
    number=str((random.randint(1, 2)))
    result_str='player'+number
    return result_str
   
```

```python
# To check if there is an empty place on the board


def space_check(board, position):
    return(board[position]==' ') 

```

```python
# To check if the board is full


def full_board_check(board):
    for i in range(1,10):
        if space_check(board,i):
            return False
        
    return True


```

```python
# Where player wants to place a marker!

def player_choice(board):
    position=0
    while position not in [1,2,3,4,5,6,7,8,9] or not space_check(board,position):
        position=int(input('Pick up a number (1-9):'))
    
    return position

```

```python
# If players want to continue playing or not!

def replay():
    
    return input('Do you want to play again? Enter Yes or No: ').lower().startswith('y')
    

```

```python
# Calling of the functions and arranging them in a sequence to build a Game!

print('Welcome to Tic Tac Toe!')

#while True:

while True:
     # Set the game up here
    the_board=[' ']*10
    player1_marker,player2_marker=player_input()
    turn=choose_first()
    print(turn+' will go first')
    play_game=input('Are you ready? Enter Yes or No:')
    if play_game.lower()[0]=='y':
        game_on =True
    else:
        game_on =False
        
       #while game_on:   
    
    while game_on:
        
        #Player 1 Turn
        
        if turn=='Player1':
            display_board(the_board)
            position=player_choice(the_board)
            place_marker(the_board, player1_marker, position)
            if win_check(the_board, player1_marker):
                display_board(the_board)
                print('Player1 won!')
                game_on=False
                
                # Check if there is a Tie between players
                
            else:
                if full_board_check(the_board):
                    display_board(the_board)
                    print('There is a Tie!')
                    game_on=False
                else:
                    turn='Player2'
                    
                    
         # Player2's turn.
        
        
        else:            
            display_board(the_board)
            position=player_choice(the_board)
            place_marker(the_board, player2_marker, position)
            if win_check(the_board, player2_marker):
                display_board(the_board)
                print('Player2 won!')
                game_on=False
            else:
                if full_board_check(the_board):
                    display_board(the_board)
                    print('There is a Tie!')
                    game_on=False
                else:
                    turn='Player1'
                    
                    
      # Play again?
    
    
    if not replay():
        break
```
